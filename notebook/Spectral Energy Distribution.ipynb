# Generate a spectral energy distribution

```python
import os
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

# Import SEDs
datafile = os.path.abspath("../SED/HS2220+2146A_SED")
names = ["wavelength","luminosity"]
SED_A = np.genfromtxt(datafile, dtype=None, usecols=range(len(names)), names=names)

datafile = os.path.abspath("../SED/HS2220+2146B_SED")
names = ["wavelength","luminosity"]
SED_B = np.genfromtxt(datafile, dtype=None, usecols=range(len(names)), names=names)
```

```python
plt.plot(SED_A["wavelength"], SED_A["luminosity"])
plt.xlim(1000.0, 15000.0)

plt.show()
```

```python
# HS 2220+2146A magnitudes
HS2220A = np.array([16.37, 16.00, 16.21, 16.46, 16.70])
# HS 2220+2146B magnitudes
HS2220B = np.array([15.65, 15.59, 15.91, 16.22, 16.50])

# Central wavelengths from Fukugita et al. (1996)
#filter_l = np.array([3557.0, 4825.0, 6261.0, 7672.0, 9097.0, 12350.0])
filter_l = np.array([3557.0, 4825.0, 6261.0, 7672.0, 9097.0])
filter_dl = np.array([463.0, 988.0, 955.0, 1064.0, 1248.0])
```

```python
lum_g = np.mean(SED_A["luminosity"][np.where(np.abs(SED_A["wavelength"] - filter_l[1])<filter_dl[1]/2.0)])

lum_u = 10**(1.0/2.5 * (HS2220A[1] - HS2220A[0]) + np.log10(lum_g))
lum_r = 10**(1.0/2.5 * (HS2220A[1] - HS2220A[2]) + np.log10(lum_g))
lum_i = 10**(1.0/2.5 * (HS2220A[1] - HS2220A[3]) + np.log10(lum_g))
lum_z = 10**(1.0/2.5 * (HS2220A[1] - HS2220A[4]) + np.log10(lum_g))
#lum_J = 10**(1.0/2.5 * (HS2220A[1] - HS2220A[5]) + np.log10(lum_g))

#A_lum = np.array([lum_u, lum_g, lum_r, lum_i, lum_z, lum_J])
A_lum = np.array([lum_u, lum_g, lum_r, lum_i, lum_z])
```

```python
plt.plot(SED_A["wavelength"], SED_A["luminosity"])

plt.scatter(filter_l, A_lum)


plt.xlim(2000.0, 15000.0)

plt.xlabel("Wavelength")
plt.tick_params(axis='y', labelleft='off')


plt.savefig("../figures/SED_HS2220A.pdf")
```

```python
lum_g = np.mean(SED_B["luminosity"][np.where(np.abs(SED_B["wavelength"] - filter_l[1])<filter_dl[1]/2.0)])

lum_u = 10**(1.0/2.5 * (HS2220B[1] - HS2220B[0]) + np.log10(lum_g))
lum_r = 10**(1.0/2.5 * (HS2220B[1] - HS2220B[2]) + np.log10(lum_g))
lum_i = 10**(1.0/2.5 * (HS2220B[1] - HS2220B[3]) + np.log10(lum_g))
lum_z = 10**(1.0/2.5 * (HS2220B[1] - HS2220B[4]) + np.log10(lum_g))

B_lum = np.array([lum_u, lum_g, lum_r, lum_i, lum_z])
```

```python
plt.plot(SED_B["wavelength"], SED_B["luminosity"])

plt.scatter(filter_l, B_lum)


plt.xlim(2000.0, 15000.0)

plt.xlabel("Wavelength")
plt.tick_params(axis='y', labelleft='off')

plt.savefig("../figures/SED_HS2220B.pdf")
```
