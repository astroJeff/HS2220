# Generate a spectral energy distribution

```python
import os
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

# Import SEDs
datafile = os.path.abspath("../SED/HS2220+2146A_SED")
names = ["wavelength","flux"]
SED_A = np.genfromtxt(datafile, dtype=None, usecols=range(len(names)), names=names)

datafile = os.path.abspath("../SED/HS2220+2146B_SED")
names = ["wavelength","flux"]
SED_B = np.genfromtxt(datafile, dtype=None, usecols=range(len(names)), names=names)

datafile = os.path.abspath("../SED/Test_SED")
names = ["wavelength","flux"]
Test = np.genfromtxt(datafile, dtype=None, usecols=range(len(names)), names=names)


# Convert from F to nu*F
Test["flux"] = Test["flux"]/Test["wavelength"] * 3.0e18
SED_A["flux"] = SED_A["flux"]/SED_A["wavelength"] * 3.0e18
SED_B["flux"] = SED_B["flux"]/SED_B["wavelength"] * 3.0e18
```

```python
#plt.plot(Test["wavelength"], Test["flux"])
plt.plot(SED_A["wavelength"], SED_A["flux"])
plt.xlim(2000.0, 25000.0)


plt.show()
```

```python
# HS 2220+2146A magnitudes
#HS2220A = np.array([16.37, 16.01, 16.21, 16.46, 16.70, 16.415])
HS2220A = np.array([16.37, 16.01, 16.21, 16.46, 16.70])
# HS 2220+2146B magnitudes
#HS2220B = np.array([15.65, 15.59, 15.91, 16.22, 16.50, 16.05])
HS2220B = np.array([15.65, 15.59, 15.91, 16.22, 16.50])

# Central wavelengths from Fukugita et al. (1996)
#filter_l = np.array([3557.0, 4825.0, 6261.0, 7672.0, 9097.0, 12350.0])
filter_l = np.array([3557.0, 4825.0, 6261.0, 7672.0, 9097.0])
filter_dl = np.array([463.0, 988.0, 955.0, 1064.0, 1248.0])
```

```python
# Fluxes from A
f_u_A = 8.91e-12
f_g_A = 9.01e-12
f_r_A = 5.73e-12
f_i_A = 3.71e-12
f_z_A = 2.46e-12
f_J_A = 1.05e-12
f_H_A = 9.61e-13
f_K_A = 3.36e-13

# Fluxes from B
f_u_B = 1.73e-11
f_g_B = 1.31e-11
f_r_B = 7.56e-12
f_i_B = 4.63e-12
f_z_B = 2.96e-12
f_J_B = 1.47e-12
f_H_B = 6.17e-13
f_K_B = 5.04e-13

# Array of Fluxes
flux_A = np.array([f_u_A,f_g_A,f_r_A,f_i_A,f_z_A,f_J_A,f_H_A,f_K_A])
flux_B = np.array([f_u_B,f_g_B,f_r_B,f_i_B,f_z_B,f_J_B,f_H_B,f_K_B])
wavelengths = np.array([359.0,481.0,623.0,764.0,906.0,1235.0,1662.0,2159.0]) * 10.0
```

```python
lum_i_A = np.mean(SED_A["flux"][np.where(np.abs(SED_A["wavelength"] - filter_l[3])<filter_dl[3]/2.0)])
factor_A = lum_i_A / flux_A[3]
print "Scaling factor for WD A:", factor_A

lum_i_B = np.mean(SED_B["flux"][np.where(np.abs(SED_B["wavelength"] - filter_l[3])<filter_dl[3]/2.0)])
factor_B = lum_i_B / flux_B[3]
print "Scaling factor for WD B:", factor_B
```

```python
plt.figure(figsize=(12,4))


plt.subplot(2,1,1)
plt.plot(SED_A["wavelength"], SED_A["flux"], color='k')
plt.scatter(wavelengths, flux_A * factor_A, color='k')
#plt.scatter(filter_l, A_lum)
plt.tick_params(axis='x', labelbottom='off')
plt.tick_params(axis='y', labelleft='off')
plt.xlim(3000.0, 15000.0)
plt.ylim(-1.0e10,1.0e11)
#plt.text(12000, 1.0e-4, "HS 2220+2146A", fontsize=14)


plt.subplot(2,1,2)
plt.rc('xtick', labelsize=14)
plt.plot(SED_B["wavelength"], SED_B["flux"], color='k')
plt.scatter(wavelengths, flux_B * factor_B, color='k')
#plt.scatter(filter_l, B_lum)
plt.xlim(3000.0, 15000.0)
plt.xlabel(r"Wavelength ($\AA$)", size=14)
plt.tick_params(axis='y', labelleft='off')
plt.ylim(-1.0e10, 2.0e11)
#plt.text(12000, 1.8e-4, "HS 2220+2146B", fontsize=14)

plt.subplots_adjust(hspace=0)

plt.show()
#plt.savefig("../figures/SED.pdf", bbox_inches='tight')
```

```python
# WISE Fluxes
f_W1 = 8.64e-14
f_W2 = 3.61e-14
f_W3 = 5.84e-14
f_W4 = 2.89e-13

# WISE upper limits
f_u_W1 = 9.30e-14
f_u_W2 = 4.46e-14
flux_WISE_err_u = np.array([f_u_W1-f_W1, f_u_W2-f_W2])
# WISE lower limits
f_l_W1 = 8.02e-14
f_l_W2 = 2.92e-14
flux_WISE_err_l = np.array([f_W1-f_l_W1, f_W2-f_l_W2])


flux_WISE = np.array([f_W1,f_W2,f_W3,f_W4])
wavelengths_WISE = np.array([34000.0, 46000.0, 120000.0, 220000.0])
```

```python
fig = plt.figure(figsize=(12,8))

ax = fig.add_subplot(1,1,1)

ax1 = fig.add_subplot(2,1,1)
ax1.plot(SED_A["wavelength"], SED_A["flux"], color='k')
ax1.scatter(wavelengths[:6], flux_A[:6] * factor_A, color='k', s=100)
ax1.errorbar(wavelengths[-2:], flux_A[-2:] * factor_A, xerr=500, yerr=1.0e10, color='k', marker='.', ls='none', uplims=True, capsize=7, elinewidth=4)
#ax1.set_xlim(3000.0, 23000.0)
ax1.set_xlim(3000.0, 50000.0)
ax1.set_ylim(-3.0e10,1.0e11)
#ax1.text(14000, 1.1e-4, "HS 2220+2146A", fontsize=30)
ax1.tick_params(axis='y', labelleft='off')
ax1.tick_params(axis='x', labelbottom='off')


ax2 = fig.add_subplot(2,1,2)
ax2.plot(SED_B["wavelength"], SED_B["flux"], color='k')
ax2.scatter(wavelengths[:6], flux_B[:6] * factor_B, color='k', s=100)
ax2.errorbar(wavelengths[-2:], flux_B[-2:] * factor_B, xerr=500, yerr=2.0e10, color='k', marker='.', ls='none', uplims=True, capsize=7, elinewidth=4)
ax2.scatter(wavelengths_WISE[:2], flux_WISE[:2] * factor_B, color='k', s=100)
#ax2.set_xlim(3000.0, 23000.0)
ax2.set_xlim(3000.0, 50000.0)
ax2.set_ylim(-3.0e10, 2.0e11)
ax2.set_xlabel(r"Wavelength ($\AA$)", size=30)
#ax2.text(14000, 2.0e-4, "HS 2220+2146B", fontsize=30)
ax2.tick_params(axis='y', labelleft='off')
ax2.tick_params(axis='x', labelsize=30)

ax.set_ylabel(r"$\lambda$ F$_{\lambda}$ (erg s$^{-1}$ cm $^{-2}$)", fontsize=30)
ax.tick_params(axis='y', labelleft='off')
ax.tick_params(axis='x', labelbottom='off')



plt.subplots_adjust(hspace=0)

#plt.savefig("../figures/SED.pdf", bbox_inches='tight')
plt.show()
```

```python
# Convert from lambda F to F, multiply by factor
F_A_SED = SED_A["flux"] / SED_A["wavelength"] * 1.0e-12
F_A_1 = flux_A[:6] * factor_A / wavelengths[:6] * 1.0e-12
F_A_2 = flux_A[-2:] * factor_A / wavelengths[-2:] * 1.0e-12
F_A_2_err = 0.8*flux_A[-2:] * factor_A / wavelengths[-2:] * 1.0e-12
F_A_3 = flux_WISE[:2] * factor_A / wavelengths_WISE[:2] * 1.0e-12
F_A_3_err_l = flux_WISE_err_l * factor_A / wavelengths_WISE[:2] * 1.0e-12
F_A_3_err_u = flux_WISE_err_u * factor_A / wavelengths_WISE[:2] * 1.0e-12

F_B_SED = SED_B["flux"] / SED_B["wavelength"] * 1.0e-12
F_B_1 = flux_B[:6] * factor_B / wavelengths[:6] * 1.0e-12
F_B_2 = flux_B[-2:] * factor_B / wavelengths[-2:] * 1.0e-12
F_B_2_err = 0.8*flux_B[-2:] * factor_B / wavelengths[-2:] * 1.0e-12
F_B_3 = flux_WISE[:2] * factor_B / wavelengths_WISE[:2] * 1.0e-12




fig = plt.figure(figsize=(12,10))

ax = fig.add_subplot(1,1,1)

ax1 = fig.add_subplot(2,1,1)
ax1.plot(SED_A["wavelength"]*1.0e-4, F_A_SED, color='k')
ax1.scatter(wavelengths[:6]*1.0e-4, F_A_1, color='k', s=100)
ax1.errorbar(wavelengths[-2:]*1.0e-4, F_A_2, xerr=0.1, yerr=F_A_2_err, color='k', marker='.', ls='none', uplims=True, capsize=7, elinewidth=4)
ax1.scatter(wavelengths_WISE[:2]*1.0e-4, F_A_3, color='k', s=100, marker='v')
#ax1.errorbar(wavelengths_WISE[:2]*1.0e-4, F_A_3, yerr=[F_A_3_err_l,F_A_3_err_u], color='k', marker='.', ls='none', markersize=1)

ax1.set_xlim(0.25000, 5.00000)
ax1.set_ylim(1.0e-9,1.0e-4)
#ax1.text(14000, 1.1e-4, "HS 2220+2146A", fontsize=30)
ax1.tick_params(axis='y', labelleft='off')
ax1.tick_params(axis='x', labelbottom='off')
ax1.set_yscale('log')





ax2 = fig.add_subplot(2,1,2)
ax2.plot(SED_B["wavelength"]*1.0e-4, F_B_SED, color='k')
ax2.scatter(wavelengths[:6]*1.0e-4, F_B_1, color='k', s=100)
ax2.errorbar(wavelengths[-2:]*1.0e-4, F_B_2, xerr=0.1, yerr=F_B_2_err, color='k', marker='.', ls='none', uplims=True, capsize=7, elinewidth=4)
ax2.scatter(wavelengths_WISE[:2]*1.0e-4, F_B_3, color='k', s=100, marker='v')
#ax2.errorbar(wavelengths_WISE[:2]*1.0e-4, F_B_3, yerr=[F_B_3_err_l,F_B_3_err_u], color='k', marker='.', ls='none', markersize=1)



ax2.set_xlim(0.25000, 5.00000)
ax2.set_ylim(2.0e-9, 5.0e-4)
ax2.set_xlabel(r"Wavelength ($\mu$m)", size=30)
#ax2.text(14000, 2.0e-4, "HS 2220+2146B", fontsize=30)
ax2.tick_params(axis='y', labelleft='off')
ax2.tick_params(axis='x', labelsize=30)
ax2.set_yscale('log')



# Create SED A subplot
ax3 = fig.add_axes([0.52,0.675,0.35,0.21],axisbg='w')
ax3.plot(SED_A["wavelength"]*1.0e-4, F_A_SED, color='k')
ax3.scatter(wavelengths[:6]*1.0e-4, F_A_1, color='k', s=100)
ax3.set_xlim(0.3,0.8)
ax3.tick_params(axis='y', labelleft='off')
ax3.set_ylim(1.0e-6,2.3e-5)
ax3.set_ylabel(r"F$_{\lambda}$", fontsize=20)
ax3.tick_params(axis='x', labelsize=20)
ax3.set_xticks([0.4,0.6,0.8])





# Create SED B subplot
ax4 = fig.add_axes([0.52,0.29,0.35,0.21],axisbg='w')
ax4.plot(SED_B["wavelength"]*1.0e-4, F_B_SED, color='k')
ax4.scatter(wavelengths[:6]*1.0e-4, F_B_1, color='k', s=100)
ax4.set_xlim(0.3,0.8)
ax4.tick_params(axis='y', labelleft='off')
ax4.set_ylim(1.0e-6,5.0e-5)
ax4.set_ylabel(r"F$_{\lambda}$", fontsize=20)
ax4.tick_params(axis='x', labelsize=20)
ax4.set_xticks([0.4,0.6,0.8])




# Create Y-axis label
ax.set_ylabel(r"log F$_{\lambda}$ (erg s$^{-1}$ cm $^{-2}$ $\AA^{-1}$)", fontsize=30)
ax.tick_params(axis='y', labelleft='off')
ax.tick_params(axis='x', labelbottom='off')



plt.subplots_adjust(hspace=0)

plt.savefig("../figures/SED.pdf", bbox_inches='tight')
#plt.show()
```
